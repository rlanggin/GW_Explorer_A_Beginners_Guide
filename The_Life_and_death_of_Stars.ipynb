# The Life and Death of Stars

In this brief notebook, we provide a quick overview of the life and death of stars!

In reality, the formation and evolution of a star depends on many factors, from the strength of interstellar magnetic fields, the abundance of chemical elements, to the presence of stellar neighbours and how fast a star is rotating. Here, we are going to ignore all of these, and focus only on one critical parameters: the **mass** of the star!

In practice, the whole life of a star can be summarised in the following:

1. A constant battle between pressure and gravity.
2. Massive stars "live big, and die young"!

But first things first, *how do we form a star*? Simple: Just bring a lot of material (gas) together! Why? So that the gravitational pull of the gas can overcome its kinetic energy (due to thermal motion). It can be shown that the gas cloud will collapse if $|U| \gtrsim 2 K$, where $U$ is the potential energy and $K$ the kinetic energy.

If $U=-\frac{3}{5} \frac{GM^2}{R}$, the gravitational potential of a spherical gas cloud, and $K=\frac{3}{2}N k_B T$, its kinetic energy due to thermal motions, with $N=M/\mu m_H$, with $\mu$ the mean molecular weight, and $m_H$ the hydrogen mass, show that a gas cloud collapses if its mass $M$ is larger than $M_J$, the so-called "*Jeans mass*". $$
M \gtrsim M_J = \left( \frac{5 k_B T}{G \mu m_H} \right)^{3/2} \left( \frac{3}{4 \pi \rho_0} \right)^{1/2},
$$

When the gas cloud collapses, different smaller regions can become locally unstable and form stars. Depending on its initial mass, with critical points at $\sim 8 M_\odot$ and $\sim 30 M_\odot$, the final outcome at the end of the star's life can be (see also Figure):

- a *White Dwarf* (WD), for $M_{\rm initial} < 8 M_\odot$.
- a *Neutron Star* (NS), for $8 M_\odot < M_{\rm initial} < 30 M_\odot$.
- a *Black Hole* (BH), for $M_{\rm initial} > 30 M_\odot$.

# Star Lives: A fight of gravity against pressure

Let's go back to pressure, and see how different (pressure) regimes affect the star's size:

Pressure is defined as a force over area, so if we take force$=$gravity, over the area of a spherical star:

$$
P \sim \frac{GM^2}{R^2}/R^2 \sim \frac{GM^2}{R^4} \sim \frac{G (\rho R^3)^2}{R^4} \sim G \rho^2 R^2,
$$

where we also introduced the density $\rho = M/V \sim M/R^3$.

### 1. Main Sequence Stars (like our Sun)

For main sequence stars, we take the source of the pressure to be the **motions of the gas**, so we use the ideal gas law:

$$
P \sim \rho \frac{T}{m} \Rightarrow T \sim \frac{Pm}{\rho} \sim \frac{GMm}{R} \propto M/R.
$$

As $R$ decreases, $T$ increases till it reaches a critical temperature for nuclear fussion to start, which is the power source of stars. Then for $T=T_{\rm crit}=$const.:

$$
{\rm const.} \sim \frac{M}{R} \Rightarrow R \sim M,\quad {\rm and}\quad \rho \sim \frac{M}{R^3} \sim 1/M^2.
$$

So for *main sequence stars*:

- Bigger mass, leads to bigger size.
- Low mass stars tend to be more dense.

---------------------

We are now turning to "*compact objects*": WDs, NSs, and BHs. We'll see why we call them "compact". The game here is the same: pressure is trying to combat gravity. As stars evolve, gradually they lose possible pressure sources till eventually, if their mass is high enough, they collapse to BHs!

## 2. White Dwarfs

The star has depleted its nuclear fuel, and cannot generate the pressure we talked about above. It starts to contract. But this is not the end: **quantum mechanics come to the rescue**! According to quantum mechanics, it is difficult to squeeze electrons together, hence a "new" form of pressure, "*electron degeneracy pressure*".

The details are a bit more difficult to follow than the case of the ideal gas law, but one case show that

$$
P \sim M^{5/3}/R^5.
$$

Let's equate this again to the inward pressure due to gravity $P \sim GM^2/R^4$ to get

$$
R \sim M^{-1/3}.
$$

For WD, the **bigger the mass, the smaller the radius** (they are compact!). For reference, a WD with $M_{\rm WD} = M_\odot$, will have a similar size with the Earth.

## 3. Neutron Stars

If the star is very massive, we have a problem. As all the material is pressured together, the electrons will be pressed against the protons of the atoms, leading to the weak interaction process $e^-+p \rightarrow n +\nu_e$. We have the production of neutrons $n$ and electron neutrinos $\nu_e$. The star blows off as a **supernova**, and what remains is a ball of neutrons.

In this situation, the neutron paly the role of the electrons before, providing the "*neutron degeneracy pressure*". This leads to the following scaling relation between the mass and radius of a NS:

$$
R \sim 14\ {\rm km}\ \left( \frac{1.4 M_\odot}{M} \right)^{1/3}.
$$

## 4. Black Holes

You guessed correctly! For even higher masses, not even the pressure of the neutrons can counteract gravity!

Moreover, there is another plot twist in the story: In relativistic scenarios, pressure can actually behave as an extra mass density, so we would have more mass to counteract. Let's see why:

Pressure is defined as force per unit area. Show that this has the same dimensions as energy per volume, i.e. $[P] = E/L^3$.

--------------------------------

From Einstein's famous equation $m=E/c^2$, mass and energy are equivalent quantities. So energy can act as an extra form of mass. If we combine this with the result of *Ex. 6*, i.e. that pressure has dimensions of energy density, we observe that it can act as a mass density. Then we enter a "cycle":

As $\rho$ increases, $P$ increases to counteract it, but the increase in $P$, in relativistic sistuations, contibutes to the mass density too. So one needs even more pressure... At some point, the pressure is defeated. The star collapses and a BH is formed!
